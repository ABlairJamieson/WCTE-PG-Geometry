```python
# From drawing of mPMT module some global constant offests
ri = gb.mpmt_dome_inner_radius/1000 #m
ro = gb.mpmt_dome_outer_radius/1000 #m
rmatrix = gb.mpmt_matrix_radius/1000 #m
rpmt = 0.04 #m gb.pmt_radius/1000 #m
print('ri=',ri, 'ro=',ro, 'rpmt=',rpmt,'rmatrix=',rmatrix,' m')
```

```python
dome_x, dome_y, dome_z, blacks_x, blacks_y, blacks_z, u_x, u_y, u_z, w_x, w_y, w_z = gb.get_mpmt_dome_blacks_pos_dir()
# convert to meters
dome_x /= 1000
dome_y /= 1000
dome_z /= 1000
blacks_x /= 1000
blacks_y /= 1000
blacks_z /= 1000

pmt_x, pmt_y, pmt_z, pmt_dirx, pmt_diry, pmt_dirz =  gb.get_pmt_loc_dir()
pmt_x /= 1000
pmt_y /= 1000
pmt_z /= 1000

cam_x, cam_y, cam_z, cam_dxx, cam_dxy, cam_dxz, cam_dzx, cam_dzy, cam_dzz = gb.get_camera_pos_dir()
cam_x /= 1000
cam_y /= 1000
cam_z /= 1000

led_x, led_y, led_z, led_ux, led_uy, led_uz, led_wx, led_wy, led_wz, led_labels = gb.get_led_loc_dir()
led_x /= 1000
led_y /= 1000
led_z /= 1000

print('lengths of arrays')
print('domes=',len(dome_x), len(dome_y), len(dome_z),'expect 9* for leds=',9*len(dome_x))
print('pmts=',len(pmt_x), len(pmt_y), len(pmt_z))
print('leds=',len(led_x), len(led_y), len(led_z),'leds/domes=',len(led_x)/len(dome_x), 'labels=',len(led_labels))
print('cams=',len(cam_x), len(cam_y), len(cam_z))
```

```python
for i in range(0,9):
    drcur = np.sqrt( (led_x[i]-led_x[0])**2 +(led_y[i]-led_y[0])**2 +(led_z[i]-led_z[0])**2 )
    print('i=',i,'drcur=',drcur)
```
